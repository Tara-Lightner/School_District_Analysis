```python
# 4.9.3 Object Oriented Programming
# Object-oriented programming (OOP) is a way of organizing information in programming. 
# The class is like the mold (to create the same type of car), and the object is like a particular toy car.
# class is used to create a class. 
#  __init__ method is a special method called a "class constructor" that Python calls every time a new instance of the class is created.
# This method takes two arguments: self and name. 
 #The self argument refers to a specific Cat object that is created when this method is called. 
# The second argument, name, will be specified each time a new Cat object is created
# an object can hold attributes, and it can also run functions.

# Example
class Cat:
    def __init__(self, name):
        self.name = name

# Here, an object whose name is Felix is created, or instantiated. 
# Since it is assigned to the variable first_cat, first_cat.name returns Felix. 
# Pay attention to the dot notation in this example. 
# The name attribute does not belong to all Cat objects; it belongs to the specific Cat object called first_cat.

first_cat = Cat('Felix')
print(first_cat.name)



# Skill Drill
# Create another object of the Cat class. Its name should be Garfield, 
# and the object should be assigned to the variable second_cat.



second_cat = Cat('Garfield')
print(second_cat.name)

# In OOP, an object can have attributes, in this case, name, color, and sound
# ??? What is self???
# such as name in the example above. It can also perform actions through functions. Let's now create a Dog class.

# It also contains a method, or a function that is associated with an object.
# ??? What is a method vs. a function ???
## Answer: When a function is associated with an object, it is called a method. If an object's attributes are ...
# ...like its adjectives and nouns ("color" and "name" for example), then its methods are like its verbs ("bark").
# This means that an object created from this Dog class can now perform an action through the bark method.
# Note the dot notation should be a clue that it is OOP

# Leason Learned: I ran this originally with both code sections below.
# I reran it and it looked like I did not need the first code section.
# However after removing the first code section, restarting and clearing the output of the kernal...
# It looks like we DO need the first section of coding (pictured directly below, so what does this represent/mean?)
# Conclusion, because 

class Dog:
    def __init__(self, name, color, sound):
        self.name = name
        self.color = color
        self.sound = sound

    def bark(self):
        return self.sound + ' ' + self.sound

# perform action though the bark method
# ??? Why does it read woolf! woolf! two times>

first_dog = Dog('Fido', 'brown', 'woof!')
print(first_dog.name)
print(first_dog.color)
first_dog.bark()

# Instantiate another object of the Dog class. Its name should be Lady, its color should be blonde, and 
# its sound should be arf!. 
# The object should be assigned to the variable second_dog. Call this object's bark method.
# Note: def sections must be indented from class to run correctly
# ??? When I run this it seems they all have to be in the same cell, but that was not true of Cat.  What is going on???


class Dog:
    def __init__(self, name, color, sound):
        self.name = name
        self.color = color
        self.sound = sound
        
    def arf(self):
        return self.sound + ' ' + self.sound

second_dog = Dog('Lady', 'blonde', 'arf!')
print(second_dog.name)
print(second_dog.color)
second_dog.arf()
```
